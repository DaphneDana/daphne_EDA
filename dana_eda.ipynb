📊 Summary of Initial Data Inspection

This section provides an overview and interpretation of the descriptive statistics generated from the dataset. The dataset contains 2,500 bank customer records with demographic information, financial attributes, and account onboarding details.

## 1. Summary of Numerical Variables
### Birth Year

All 2,500 entries have valid birth years.

The average birth year is 1989, meaning the typical customer is around 36 years old.

Ages range from approximately 20 years (born 2005) to 67 years (born 1958).

The distribution appears realistic and well-behaved with no implausible values.

Implication:
The age range is diverse but valid, providing good demographic coverage without requiring corrections.

### Monthly Income (UGX)

All entries contain income values.

The average monthly income is UGX 846,337, with a median of UGX 434,760, indicating right-skewness.

The standard deviation (UGX 2.5M) is extremely high, suggesting significant income variability.

The maximum income (UGX 91M) is an extreme outlier, far above typical Ugandan income ranges.

Implication:
Income data is highly skewed with substantial outliers. Outlier detection and possible capping or transformation will be necessary during preprocessing.

### KYC Level

Out of 2,500 entries, 2,378 have valid KYC values, meaning 122 are missing.

The average KYC level is 1.64, with valid levels ranging from 1 to 3.

Distribution suggests most customers fall under KYC levels 1 and 2.

Implication:
KYC values are straightforward but missing values require imputation (likely using the median value).

### Smartphone Ownership

2,381 customers have valid smartphone status values → 119 missing.

73% of customers own a smartphone.

Binary distribution (0/1) is clean and interpretable.

Implication:
Missing values will need imputation. The mode (1) is most appropriate since smartphone ownership is common.

## 2. Summary of Categorical Variables
### Customer ID

All 2,500 IDs are unique, confirming no duplicate customer records.

### District

Data spans 60 districts, with Kampala having the highest representation (464 customers).

Implication:
Strong urban concentration, especially around Kampala.

### Region

Four valid regions appear: Central, Eastern, Western, Northern.

Central region dominates with 1,285 customers (51%).

Implication:
The dataset may reflect customer concentration in more urbanized regions.

### Employment Status

There are 8 employment categories, with self-employed being the most common (328 customers).

Implication:
Employment categories appear reasonable but may require text normalization for consistency.

### Phone Carrier

Only 3 carriers appear (MTN, Airtel, Africell), with MTN being most common (59%).

Implication:
Carrier distribution aligns with Uganda’s market share. Minor text cleaning may be necessary.

### Date Joined

Contains 1,174 unique joining dates, with some dates repeated up to 7 times.

Implication:
Column looks structurally correct but must be converted to actual datetime. Future dates should be validated.

## 3. Key Issues Identified for Cleaning

Missing Values

KYC Level (122 missing)

Smartphone Ownership (119 missing)

Outliers

Monthly income has extremely high values (e.g., 91M UGX).

Data Type Fixes Needed

date_joined → convert to datetime

has_smartphone → convert to boolean/int

kyc_level, birth_year, monthly_income_ugx → ensure numeric

Categorical Cleanup

Standardize text values (e.g., region names, phone carriers, employment statuses)

Possible Feature Engineering

Age (from birth_year)

Customer tenure (from date_joined)

## 4. Overall Interpretation

The dataset is generally well-structured and rich in useful customer-level information. Most numeric and categorical values fall within realistic ranges, and customer IDs are completely unique. However, several columns require additional cleaning, mainly around missing values, outlier handling, and type conversion. Addressing these issues will prepare the dataset for meaningful exploratory analysis and profiling.

```python
# Count of missing values per column
df.isna().sum()
```

```python
# Percentage of missing values per column
(df.isna().sum() / len(df)) * 100
```

```python
missing_summary = (
    df.isna().sum()
      .to_frame(name='missing_count')
      .assign(missing_pct=lambda x: x['missing_count'] * 100 / len(df))
)
missing_summary
```

```python
# Impute KYC level with median
df['kyc_level'] = df['kyc_level'].fillna(df['kyc_level'].median())
```

```python
# Impute smartphone ownership with mode
df['has_smartphone'] = df['has_smartphone'].fillna(df['has_smartphone'].mode()[0])
```
